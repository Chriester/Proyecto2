```python
def categorize_time_of_day(hour):
    if 7 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 17:
        return 'afternoon'
    elif 17 <= hour < 21:
        return 'evening'
    else:
        return 'night'
```

```python
datos['franja_dia'] = datos['created_at'].dt.hour.apply(categorize_time_of_day)
datos['hora_creacion'] = datos['created_at'].dt.hour
datos_franja_dia = datos.groupby('franja_dia').size().sort_values()
```

```python
# Calcular el rango intercuartílico (IQR)
Q1 = datos['amount'].quantile(0.25)  # Primer cuartil
Q3 = datos['amount'].quantile(0.75)  # Tercer cuartil
IQR = Q3 - Q1  # Rango intercuartílico

# Determinar los límites para los outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar outliers
outliers = datos[(datos['amount'] < lower_bound) | (datos['amount'] > upper_bound)]

# Filtrar los datos por status
outliers_filtered = outliers[outliers['status'].isin(['rejected','money_back','canceled','active','direct_debit_rejected','transaction_declined','direct_debit_sent'])]

# Mostrar los resultados
print(outliers['status'].value_counts())
print(outliers_filtered[['amount','status']])
print(datos['status'].unique())
'''
Todos los outliers cumplen las siguientes condiciones
-el amount es 200
-El status es rejected o money_back
-el transfer type es regular
'''


```

### Se suma "amount"

```python
# Realiza el merge entre `df_fees` y los datos necesarios de `datos`
df_merge = df_fees.merge(
    datos[['id', 'user_id', 'amount']],  # Se incluye la columna "amount" de `datos`
    left_on='cash_request_id',
    right_on='id',
    suffixes=('', '_cash')
)

df_merge = df_merge.dropna(subset=['amount', 'cash_request_id', 'user_id'])

# Elimina duplicados en las transacciones basándonos en 'user_id' y 'cash_request_id'
df_merge = df_merge.drop_duplicates(subset=['user_id', 'cash_request_id'])

# Calcula métricas por cliente, usando la columna "amount" de `datos`
clientes = df_merge.groupby('user_id').agg(
    total_monto=('amount', 'sum'),  # Sumar la columna "amount" de `datos`
    transacciones=('cash_request_id', 'count')  # Contar las transacciones
).reset_index()

# Elimina duplicados en las métricas calculadas
clientes = clientes.drop_duplicates(subset=['total_monto', 'transacciones'])

# Crea segmentos basados en percentiles de "total_monto"
bins_monto = pd.qcut(clientes['total_monto'], q=4, duplicates='drop', retbins=True)[1]
clientes['segmento_monto'] = pd.qcut(
    clientes['total_monto'],
    q=len(bins_monto) - 1,
    labels=['Básico', 'Bronce', 'Premium', 'Platinum'][:len(bins_monto) - 1]
)

# Crea segmentos basados en percentiles de "transacciones"
bins_trans = pd.qcut(clientes['transacciones'], q=4, duplicates='drop', retbins=True)[1]
clientes['segmento_transacciones'] = pd.qcut(
    clientes['transacciones'],
    q=len(bins_trans) - 1,
    labels=['Básico', 'Bronce', 'Premium', 'Platinum'][:len(bins_trans) - 1]
)

# Une los segmentos
clientes['segmento_final'] = clientes.apply(
    lambda x: f"{x['segmento_monto']}-{x['segmento_transacciones']}" 
    if not pd.isnull(x['segmento_monto']) and not pd.isnull(x['segmento_transacciones']) 
    else 'Sin Clasificar',
    axis=1
)

print(clientes)
```
